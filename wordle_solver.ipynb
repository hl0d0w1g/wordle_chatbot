```python
from spellchecker import SpellChecker
import gradio as gr
from collections import Counter
import spacy
```

```python
spanish = SpellChecker(language='es')

nlp = spacy.load('es_core_news_sm')
```

```python
spanish_corpus = dict(spanish.word_frequency.dictionary)
print('Corpus lenght:', len(spanish_corpus))
spanish_corpus
```

```python
spanish_corpus_clean = {word: freq for word, freq in spanish_corpus.items() if len(word) == 5}
spanish_corpus_clean = dict(sorted(spanish_corpus_clean.items(), key=lambda item: item[1], reverse=True))
print('Corpus lenght:', len(spanish_corpus_clean))
spanish_corpus_clean
```

```python
words_to_save = spanish_corpus_clean.copy()
words_to_save = [w for w, f in words_to_save.items() if nlp(w)[0].pos_ == 'NOUN']

with open('spanish_words.txt', 'a') as f:
    for w in words_to_save:
        f.write(str(w) + '\n')
```

```python
spanish_corpus_filtered = spanish_corpus_clean.copy()

def word_matcher(rules, corpus):
    '''
    - rules (list): [('h', 0), ('a', 0), ('c', 1), ('e', 0), ('r', 2)] 
    #### 1: The letter is not present in the word
    #### 1: The letter is present in the word but in uncorrect position
    #### 2: The letter is present in the word and in correct position
    '''
    assert(len(rules) == 5)

    for idx, (letter, rule) in enumerate(rules):

        if rule == 0:
            corpus = {word: freq for word, freq in corpus.items() if letter.lower() not in word.lower()}
        
        elif rule == 1:
            corpus = {word: freq for word, freq in corpus.items() if letter.lower() in word.lower()}

        elif rule == 2:
            corpus = {word: freq for word, freq in corpus.items() if letter.lower() == word.lower()[idx]}

        elif rule == -1:
            # Repeated letter
            pass
        else:
            raise ValueError()

    return corpus

def check_repeated_letters(rules):
    '''

    '''
    word = [l for l, r in rules]
    repeated_letters = dict(Counter(word))

    repeated_letters_rules = []
    for letter, n in repeated_letters.items():
        if n > 1:
            repeated_letters_rules.append([(idx, l, r) for idx, (l, r) in enumerate(rules) if l == letter])                

    for repeated in repeated_letters_rules:
        max_ = max(repeated, key = lambda t: t[2])
        rules = [(l, -1) if l == max_[1] and idx != max_[0] else (l, r) for idx, (l, r) in enumerate(rules)]

    return rules

def matcher(word, r1, r2, r3, r4, r5):
    '''

    '''
    assert(len(word) == 5)

    rules_encoder = {'GRIS': 0, 'AMARILLO': 1, 'VERDE': 2}
    rules = [(word[0], r1), (word[1], r2), (word[2], r3), (word[3], r4), (word[4], r5)]
    rules = [(l.lower(), rules_encoder[r]) for l, r in rules]

    rules = check_repeated_letters(rules)

    global spanish_corpus_filtered
    spanish_corpus_filtered = word_matcher(rules, spanish_corpus_filtered)
    most_probable_word = str(list(spanish_corpus_filtered.keys())[0])

    # print('Remaining words:', len(spanish_corpus_filtered))
    # print(spanish_corpus_filtered['noche'])

    spanish_corpus_filtered.pop(most_probable_word)
    return most_probable_word
```

```python
iface = gr.Interface(
  fn=matcher, 
  inputs=[
          gr.inputs.Textbox(lines=1, placeholder=None, default='', label='Palabra'),
          gr.inputs.Radio(['GRIS', 'AMARILLO', 'VERDE'], type='value', default=None, label='Regla 1'),
          gr.inputs.Radio(['GRIS', 'AMARILLO', 'VERDE'], type='value', default=None, label='Regla 2'),
          gr.inputs.Radio(['GRIS', 'AMARILLO', 'VERDE'], type='value', default=None, label='Regla 3'),
          gr.inputs.Radio(['GRIS', 'AMARILLO', 'VERDE'], type='value', default=None, label='Regla 4'),
          gr.inputs.Radio(['GRIS', 'AMARILLO', 'VERDE'], type='value', default=None, label='Regla 5')
  ], 
  outputs=[
          gr.outputs.Textbox(type='auto', label='Respuesta más probable')
  ]
)
```
